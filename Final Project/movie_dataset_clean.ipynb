```python
import pandas as pd
import json

# Load the existing CSV dataset into a DataFrame
existing_data = pd.read_csv('/Users/camellia/Documents/GitHub/BIO174-2024/AC-homework/Final Project/tmdb_5000_movies.csv')

# Function to extract values after "name":
def extract_names(json_str):
    try:
        names = [item['name'] for item in json.loads(json_str)]
        return ", ".join(names)
    except (json.JSONDecodeError, TypeError):
        return ""

# Apply the function to genres and keywords columns
existing_data['genres'] = existing_data['genres'].apply(extract_names)
existing_data['keywords'] = existing_data['keywords'].apply(extract_names)

# Selecting specific columns
selected_columns = ['budget', 'genres', 'keywords', 'original_language', 'overview', 
                    'popularity', 'release_date', 'title', 'vote_average']

# Create a new DataFrame with selected columns
new_dataframe = existing_data[selected_columns]

# Display the new DataFrame
print(new_dataframe)
new_dataframe.to_csv('new_dataset.csv', index=False)
```
