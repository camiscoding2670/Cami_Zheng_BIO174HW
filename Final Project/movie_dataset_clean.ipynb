```python
import pandas as pd
import json

# Load the existing CSV dataset into a DataFrame
existing_data = pd.read_csv('tmdb_5000_movies.csv')

# Function to extract values after "name":
def extract_names(json_str):
    try:
        names = [item['name'] for item in json.loads(json_str)]
        return ", ".join(names)
    except (json.JSONDecodeError, TypeError):
        return ""
    
# Function to extract the year from release_date:
def extract_year(date_str):
    try:
        return date_str.split('-')[0]
    except AttributeError:
        return ""
    
# Extract data
existing_data['genres'] = existing_data['genres'].apply(extract_names)
existing_data['keywords'] = existing_data['keywords'].apply(extract_names)
existing_data['production_companies'] = existing_data['production_companies'].apply(extract_names)
existing_data['release_date'] = existing_data['release_date'].apply(extract_year)

# Selecting specific columns
selected_columns = ['genres', 'homepage','keywords', 'original_language', 'overview', 
                    'popularity', 'production_companies','release_date', 'title', 'vote_average']

# Create a new DataFrame with selected columns
new_dataframe = existing_data[selected_columns]

# Display the new DataFrame
print(new_dataframe)
new_dataframe.to_csv('new_dataset.csv', index=False)
```
