```python
#!pip install svgwrite

from IPython.display import SVG, display
import svgwrite

def draw_happy_house():
    dwg = svgwrite.Drawing('happy_house.svg', profile='tiny')

    # Blue sky
    dwg.add(dwg.rect(insert=(0, 0), size=('100%', '150'), fill='skyblue'))


    # Sun on the right corner
    dwg.add(dwg.circle(center=(380, 20), r=20, fill='yellow'))

    # House
    dwg.add(dwg.rect(insert=(150, 100), size=(100, 100), fill='#ffcc66'))
    dwg.add(dwg.rect(insert=(130, 100), size=(150, 10), fill='#996633'))
    dwg.add(dwg.polygon(points=[(130, 100), (200, 50), (280, 100)], fill='#996633'))

    # Happy people on the land with horizontal smiles
    dwg.add(dwg.circle(center=(170, 160), r=5, fill='#00cc00'))
    dwg.add(dwg.circle(center=(190, 160), r=5, fill='#00cc00'))
    dwg.add(dwg.line(start=(170, 170), end=(190, 170), stroke='#00cc00', stroke_width=2))
    
    # Draw happy faces (eyes and mouth)
    dwg.add(dwg.text("..", insert=(167, 162), font_size="4", fill="black"))
    dwg.add(dwg.text("..", insert=(188, 162), font_size="4", fill="black"))
    dwg.add(dwg.text("_)", insert=(177, 175), font_size="5", fill="black"))

    # Grass
    dwg.add(dwg.rect(insert=(0, 200), size=('100%', 10), fill='#00cc00'))

    # Flowers
    flower_positions = [(50, 250), (100, 250), (150, 250)]
    for x, y in flower_positions:
        dwg.add(dwg.circle(center=(x, y), r=5, fill='#ff99cc'))
        dwg.add(dwg.line(start=(x, y), end=(x, y-20), stroke='green', stroke_width=2))
        dwg.add(dwg.line(start=(x, y), end=(x-10, y-10), stroke='green', stroke_width=2))
        dwg.add(dwg.line(start=(x, y), end=(x+10, y-10), stroke='green', stroke_width=2))

    display(SVG(dwg.tostring()))

draw_happy_house()
```

```python
## Scatter Plot Program
def myScatter(x,y):
    '''Function to draw a scatter plot in SVG
    Input: two lists with the same size
    Output: String containing the scatter plot.'''
    #Set some of the SVG properties
    svgWidth = 500
    svgHeight = 500
    pointColor = "green"
    xmargin = 50 # x-coordinate of the y-axis
    ymargin = 50 # y-coordinate of the x-axis
    xaxis = 400 # width of x-axis
    yaxis = 400 # height of y-axis
    xLeft = xmargin
    xRight = xmargin + xaxis
    yBottom = svgHeight - ymargin
    yTop = svgHeight - ymargin - yaxis

    #Let's test if the lists have the same size, if not return false
    if len(x) != len(y):
        return 0
    #if yes find the maximum and the minimum x and y
    xmin = min(x)
    ymin = min(y)

    xmax = max(x)
    ymax = max(y)

    xRange = xmax - xmin   
    yRange = ymax - ymin

    #let's create a string with the SVG code
    #we will create the canvas and draw the axes
    svg = '''<svg width="{width}" height="{height}" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink">
<line x1="{xleft}" y1="{ybottom}" x2="{xright}" y2="{ybottom}" stroke="black" stroke-width="2" />
<line x1="{xleft}" y1="{ybottom}" x2="{xright}" y2="{ybottom}" stroke="black" stroke-width="2" />
<line x1="{xleft}" y1="{ybottom}" x2="{xleft}" y2="{ytop}" stroke="black" stroke-width="2" />'''.format(width=svgWidth, height=svgHeight, 
                    xleft=xLeft, ybottom=yBottom, xright=xRight, ytop=yTop)

    #let's draw the axes labels
    svg = svg + '''
<text x="{xleft}" y="{ytop}" font-family="Verdana" font-size="14" fill="red">({xmin},{ymax})</text>
<text x="{xleft}" y="{ybottom}" font-family="Verdana" font-size="14" fill="red">({xmin},{ymin})</text>
<text x="{xright}" y="{ybottom}" font-family="Verdana" font-size="14" fill="red">({xmax},{ymin})</text>
    '''.format(xleft=xLeft-40, xright=xRight, ytop=yTop-20, ybottom=yBottom+20, xmax=xmax, 
                    xmin=xmin, ymax=ymax, ymin=ymin)

    #add the points to the graph. 
    for xi,yi in zip(x,y):
        svg = svg + '''
<circle cx="{xcoord}" cy="{ycoord}" r="5" fill="blue"></circle>'''.format(
        xcoord=xLeft+((xaxis/xRange)*(xi-xmin)), 
        ycoord=yBottom-((yaxis/yRange)*(yi-ymin)))

     # Add the line to the graph.
    svg = svg + f'''
    <polyline points="'''
    for xi, yi in zip(x, y):
        svg = svg + f'{xLeft + ((xaxis / xRange) * (xi - xmin))},{yBottom - ((yaxis / yRange) * (yi - ymin))} '
    svg = svg + f'''" fill="none" stroke="green" stroke-width="2" />
    '''

    #close the svg element
    svg = svg + '''
    </svg> '''

    return svg

#create some random data to plot
import random
import random
x = [random.randint(0,1000) for r in range(300)]
y = [random.randint(0,1000) for r in range(300)]

svg = myScatter(x, y)

#fileout = open("test.svg","w")
#fileout.write(svg)
#fileout.close()

from IPython.display import SVG
SVG(data=svg)
```
